```python
import pandas as pd
```

```python
data_outc = pd.read_csv('/Users/sowmy/Documents/JigsawAcad/CapstonProject_1/outcomes.csv/outcomes.csv')
```

```python
data_outc.head()
```

```python
data_outc.shape
```

```python
do = data_outc.describe(include='all')
```

```python
pd.DataFrame(do)
```

```python
do = do.T
```

```python
do.shape
```

```python
miss = data_outc.isnull().sum()
miss
```

```python
do['missing'] = miss
```

```python
do.head()
```

```python
miss_perc = data_outc.isnull().sum()/len(data_outc)*100
miss_perc
```

```python
do['missing_perc'] = miss_perc
```

```python
q5 = data_outc.quantile(0.05)
q5
```

```python
do['q5'] = q5
```

```python
q10 = data_outc.quantile(0.1)
q10
```

```python
do['q10'] = q10
```

```python
q85 = data_outc.quantile(0.85)
q85
```

```python
do['q85'] = q85
```

```python
q95 = data_outc.quantile(0.95)
q95
```

```python
do['q95'] = q95
```

```python
q99 = data_outc.quantile(0.99)
q99
```

```python
do['q99'] = q99
```

```python
do.head()
```

```python
obs = data_outc.shape[0]
```

```python
do['obs'] = obs
```

```python
do = do.reindex(columns = ['mean','std','count','obs','missing', 'missing_perc', 'min', 'max', 'unique', 'q5', 'q10', '25%', 
                         '50%', '75%', 'q85', 'q95', 'q99'])
```

```python
do.fillna('NotApplicable', inplace=True)
```

```python
do.head()
```

```python
do.to_csv('outcome_DQR.csv')
```
